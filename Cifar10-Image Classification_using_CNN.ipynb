# Step 3-Data Preparation

```python
X_train.dtype
```

```python
X_train
```

```python
X_train=X_train.astype('float32')
X_test=X_test.astype('float32')
```

```python
X_train
```

```python
import keras
num_cat=10
y_train=keras.utils.to_categorical(y_train,num_cat)
y_test=keras.utils.to_categorical(y_test,num_cat)
```

```python
y_train
```

```python
X_train=X_train/255
X_test=X_test/255
X_train
```

```python
Input_shape=X_train.shape[1:]
Input_shape
```

# Step 4- Model training

```python
from keras.models import Sequential
from keras.layers import Conv2D,MaxPooling2D,AveragePooling2D,Dense,Flatten,Dropout
from keras.optimizers import Adam
from keras.callbacks import TensorBoard
```

```python
cnn_model=Sequential()
cnn_model.add(Conv2D(filters=128,kernel_size=(3,3),activation='relu',input_shape=Input_shape))
cnn_model.add(Conv2D(filters=128,kernel_size=(3,3),activation='relu'))
cnn_model.add(MaxPooling2D(2,2))
cnn_model.add(Dropout(0.3))
```

```python
cnn_model.add(Conv2D(filters=256,kernel_size=(3,3),activation='relu'))
cnn_model.add(Conv2D(filters=256,kernel_size=(3,3),activation='relu'))
cnn_model.add(MaxPooling2D(2,2))
cnn_model.add(Dropout(0.2))
```

```python
cnn_model.add(Flatten())
cnn_model.add(Dense(units=1024,activation='relu'))
cnn_model.add(Dense(units=1024,activation='relu'))
cnn_model.add(Dense(units=10,activation='softmax'))
```

```python
cnn_model
```

```python
from tensorflow import keras
from tensorflow.keras import optimizers
```

```python
cnn_model.compile(loss='categorical_crossentropy',optimizer=keras.optimizers.RMSprop(learning_rate=0.001),metrics=['accuracy'])
```

```python
history=cnn_model.fit(X_train,y_train,batch_size=32,epochs=10,shuffle=True)
```

# Step 5- Model Evaluation

```python
evaluation=cnn_model.evaluate(X_test,y_test)
```

# Step 7- Model training using Augmented Dataset

```python
import keras
from keras.datasets import cifar10
(X_train,y_train),(X_test,y_test)=cifar10.load_data()
```

```python
from keras.preprocessing.image import ImageDataGenerator

datagen=ImageDataGenerator(rotation_range=90,
                           width_shift_range=0.1,
                           horizontal_flip=True,
                           vertical_flip=True)
datagen.fit(X_train)
```

```python
num_cat=10
y_train=keras.utils.to_categorical(y_train,num_cat)
y_test=keras.utils.to_categorical(y_test,num_cat)
```

```python
y_test.shape
```

```python
y_train.shape
```

```python
X_train=X_train.astype('float32')
X_test=X_test.astype('float32')
```

```python
X_train=X_train/255
X_test=X_test/255
```

```python
cnn_model.fit(datagen.flow(X_train,y_train,batch_size=32),epochs=10)
```

```python
score=cnn_model.evaluate(X_test,y_test)
print('Test accuracy',score[1])
```

```python
directory = os.path.join(os.getcwd(), 'Created-models')

if not os.path.isdir(directory):
    os.makedirs(directory)
model_path = os.path.join(directory, 'cifar10_images_trained_model_cnn_Augmentation.h5')
cnn_model.save(model_path)
```
